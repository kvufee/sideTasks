```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import binom
```

## Подсчитываем биномиальное распределение

```python
amnt = 100
prob = 0.6

data = binom.rvs(amnt, prob, size=1000)
```

## Строим график PDF

```python
sns.histplot(data, kde = True);
```
